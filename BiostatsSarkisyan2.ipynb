### Sample 1D density by the AR algorithm.

$$ p(x) \propto -2x^2 + 5x + 3$$ , $$ 0 \leq x \leq 3$$

#### 1. 
First, we have to find a normalizing constant so that

$$ \int_{a}^{b}p(x)dx = 1$$

Let's find that constant


$$ \int_{0}^{3}A(-2x^2 + 5x + 3)dx = 1$$

$$ A \frac{27}{2} = 1  \Rightarrow A = \frac{2}{27}$$

Now we have to accept points that are above the graph of the function and reject that are below:

$$ f(x) = A(-2x^2 + 5x + 3) = -\frac{4}{27}x^2 + \frac{10}{27}x + \frac{2}{9}$$

$$ x_{max} = \frac{5}{4} \Rightarrow f(x_{max}) = f(\frac{5}{4}) = \frac{49}{108}$$

## Ploting the histogram

### N = 1000

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
import time

a = 0.  # xmin
b = 3.  # xmax

m = 49/108 # ymax

def f(x):
    return 2*(-2*x**2 + 5*x + 3)/27  # probability density function

accept = []
reject = []
yacc = []
yrej = []
N = 1000
u1 = np.random.uniform(a, b, N)
#u2 = np.random.uniform(0.0, m, N)
for i in range(N):
#    u1 = np.random.uniform(a, b, N)
    u2 = np.random.uniform(0.0, m)

    if u2 <= f(u1[i]):
        accept.append(u1[i])
        yacc.append(u2)
    else:
        reject.append(u1[i])
        yrej.append(u2)



print("Rejections: ", len(reject))

x = np.linspace(a, b, 1000)
plt.hist(accept, 40, ec = 'black', density=True)
plt.plot(x, f(x), linewidth = '3')
plt.title('N = 1000')
plt.show()

```

#### Orange curv is our function.

```python
ss.probplot(accept, plot=plt) 
plt.show()
```

```python
x = np.linspace(a, b, 1000)

acc = np.array(accept)
rej = np.array(reject)
plt.plot(acc, yacc, '.', c = 'g')
plt.plot(rej, yrej, 'x', c = 'r')
plt.plot(x, f(x), linewidth = '5')


plt.show()
```

Green dots are accepted points and red crosses are rejected points

As shown in the histogram, there are 251 rejected points and the method does not work well with a small amount of data. We can increase the number of points, thereby improving the accuracy of the method.
